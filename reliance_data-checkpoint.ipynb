```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x_train = ss.fit_transform(x_train)
x_test = ss.fit_transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
linear_regressor=LinearRegression()

linear_regressor.fit(x_train,y_train)
```

```python
print("Accuracy of LR on training set:",linear_regressor.score(x_train,y_train))
print("Accuracy of LR on testing set:",linear_regressor.score(x_test,y_test))
```

```python
y_pred_lr=linear_regressor.predict(x_test)
y_pred_lr
```

```python
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse

print("---linear regression - model evaluation ---")
print("mean absolote error(MAE):",mae(y_test,y_pred_lr))
print("mean Squared error(MSE):",mse(y_test,y_pred_lr))
print("Root mean squared error(RMSE):",np.sqrt(mse(y_test,y_pred_lr)))
```

```python
#decision tree algorithm 

from sklearn.tree import DecisionTreeRegressor
decision_regressior=DecisionTreeRegressor()
decision_regressior.fit(x_train,y_train)
```

```python
print("Accuracy of LR on training set:",decision_regressior.score(x_train,y_train))
print("Accuracy of LR on testing set:",decision_regressior.score(x_test,y_test))
```

```python
y_pred_lr=decision_regressior.predict(x_test)
y_pred_lr
```

```python
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse

print("---decition tree - model evaluation ---")
print("mean absolote error(MAE):",mae(y_test,y_pred_lr))
print("mean Squared error(MSE):",mse(y_test,y_pred_lr))
print("Root mean squared error(RMSE):",np.sqrt(mse(y_test,y_pred_lr)))
```

```python
from sklearn.ensemble import RandomForestRegressor

random_forest_regressor=RandomForestRegressor()
random_forest_regressor.fit(x_train,y_train)
```

```python
print("Accuracy of LR on training set:",random_forest_regressor.score(x_train,y_train))
print("Accuracy of LR on testing set:",random_forest_regressor.score(x_test,y_test))
```

```python
y_pred_lr=random_forest_regressor.predict(x_test)
y_pred_lr
```

```python
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse

print("---random forest - model evaluation ---")
print("mean absolote error(MAE):",mae(y_test,y_pred_lr))
print("mean Squared error(MSE):",mse(y_test,y_pred_lr))
print("Root mean squared error(RMSE):",np.sqrt(mse(y_test,y_pred_lr)))
```

```python
import numpy as np
import joblib


def predict_score(
    Prev_Close=242, Open=292, High=346, Low=309, Last=119, Close=978,
    VWAP=331, Volume=5952000, Turnover=15000000, Trades=50217, 
    Deliverable_Volume=1766934, percent_deliverable=25  # Retained but excluded in array
):
    temp_array = [
        Prev_Close, Open, High, Low, Last, Close, VWAP, Volume,
        Turnover, Trades, Deliverable_Volume  
    ]
    
    # Convert to NumPy array and reshape
    temp_array = np.array([temp_array])  # Ensure input is 2D
    
    # Log the input array
    print("Input Array:", temp_array)
    
    # Check if input shape matches model requirements
    if temp_array.shape[1] != random_forest_regressor.n_features_in_:
        raise ValueError(
            f"Input data has {temp_array.shape[1]} features, but the model expects {random_forest_regressor.n_features_in_} features."
        )
    
    # Predict using the trained model
    predicted_value = random_forest_regressor.predict(temp_array)
    return int(predicted_value[0])

```
